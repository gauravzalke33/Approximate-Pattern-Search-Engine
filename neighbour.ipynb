## Immediate neighbour

- You start with a string Pattern (like "ACG").

- You keep the original pattern itself in the result.

- Then, for each position in the pattern:

- Replace the letter there with the other three possible nucleotides (A, C, G, T except the original one).

- Add these new strings to your set.

- Return the set.

```python
def immediateneighbour(pattern):
    nucleotides = ["A","T","G","C"]
    n = len(pattern)
    neighbourhood = set()
    for i in range(len(pattern)):
        symbol = pattern[i]
        for x in nucleotides:
            if symbol != x:
                neighbour = pattern[:i] + x + pattern[i+1:]
                neighbourhood.add(neighbour)
    return neighbourhood

immediateneighbour("AA")
```

this algorithm is to get the all possible patterns from the given pattern(with only one hamming distance) to get the immidiate neighbour

```python
def hammingdistance(p,q):
    distance = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            distance += 1
    return distance
def neighbour(pattern,d):
    nucleotides = ["A","T","G","C"]

    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return set(nucleotides)
    
    neighbourhood = set()
    suffix = pattern[1:]
    suffixneighbours = neighbour(suffix,d)

    for text in suffixneighbours:
        if hammingdistance(text,suffix) < d:
            for x in nucleotides:
                neighbourhood.add(x+text)
        else:
            neighbourhood.add(pattern[0]+text)
        
    return neighbourhood

# neighbour("AA",1)
print(len(neighbour("CCAGTCAATG",1)))
```

```python
def HammingDistance(p, q):
    return sum(pc != qc for pc, qc in zip(p, q))

def Neighbors(Pattern, d):
    nucleotides = ['A', 'C', 'G', 'T']

    if d == 0:
        return {Pattern}
    if len(Pattern) == 1:
        return set(nucleotides)

    Neighborhood = set()
    suffix = Pattern[1:]
    SuffixNeighbors = Neighbors(suffix, d)

    for text in SuffixNeighbors:
        if HammingDistance(suffix, text) < d:
            for x in nucleotides:
                Neighborhood.add(x + text)
        else:
            Neighborhood.add(Pattern[0] + text)

    return Neighborhood

print(*Neighbors("TCCAATGAG",3),sep=" ")
```
